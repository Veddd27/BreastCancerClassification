```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#Loading the dataset from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
BreastCancer_df = pd.DataFrame(breast_cancer_dataset.data , columns = breast_cancer_dataset.feature_names)
```

```python
BreastCancer_df.head()
```

```python
BreastCancer_df.shape
```

```python
BreastCancer_df["label"]=breast_cancer_dataset.target
```

```python
BreastCancer_df.tail()
```

```python
BreastCancer_df.info()
```

```python
BreastCancer_df.describe()
```

```python
x = BreastCancer_df.drop(columns = "label", axis = 1)
y =  BreastCancer_df["label"]
```

```python
print(x.shape, y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_train_std = scalar.fit_transform(x_train)
x_test_std = scalar.transform(x_test)
```

```python
print(x_train_std, x_test_std)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20, activation = "relu"),
                        keras.layers.Dense(2, activation = "sigmoid")
])
```

```python
model.compile(optimizer="adam",
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=15)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training data', 'Validation Data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Accuracy")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training data', 'Validation Data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```
